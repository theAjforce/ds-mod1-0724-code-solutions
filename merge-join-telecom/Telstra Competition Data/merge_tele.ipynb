```python
import pandas as pd
from functools import reduce
```

```python
entry_data = pd.read_csv("event_type.csv")
log_data = pd.read_csv("log_feature.csv")
resource_data = pd.read_csv("resource_type.csv")
sample_data = pd.read_csv("sample_submission.csv")
severity_data = pd.read_csv("severity_type.csv")
test_data = pd.read_csv("test.csv")
train_data = pd.read_csv("train.csv")
datas = [entry_data, log_data, resource_data, severity_data, test_data, train_data]
```

```python
train_data
```

```python
first_merge = pd.merge(entry_data, log_data, on='id', how= 'inner')
first_merge
```

```python
second_merge = pd.merge(first_merge, resource_data, on='id', how='inner')
second_merge
```

```python
third_merge = pd.merge(second_merge, sample_data, on='id', how='inner')
third_merge
```

```python
fourth_merge = pd.merge(third_merge, severity_data, on='id', how='inner')
fourth_merge
```

```python
fifth_merge = pd.merge(fourth_merge, test_data, on='id', how='inner')
fifth_merge
```

```python
sixth_merge = pd.merge(fifth_merge, train_data, on='location', how='inner')
sixth_merge
```

```python
df_almost = sixth_merge.drop(['predict_0', 'predict_1', 'predict_2', 'id_y'], axis=1)
```
